```python
from abc import ABC, abstractmethod

#Kalıtım ve Kapsülleme

class Kaynak(ABC):

    def __init__(self, baslik, kayitNo):
        self._baslik = baslik  # Kapsülleme için _ private yapısı
        self._kayitNo = kayitNo

    @property
    def baslik(self):
        return self._baslik

    @baslik.setter
    def baslik(self, value):
        if value.strip():
            self._baslik = value

    @property
    def kayitNo(self):
        return self._kayitNo

    @kayitNo.setter
    def kayitNo(self, value):
        if value.strip():
            self._kayitNo = value

    @abstractmethod
    def __str__(self):
        pass


class Kitap(Kaynak):
    """Kaynak sınıfından türetilen Kitap sınıfı."""
    def __init__(self, baslik, kayitNo, yazar, sayfa_sayisi):
        super().__init__(baslik, kayitNo)
        self._yazar = yazar
        self._sayfa_sayisi = sayfa_sayisi

    @property
    def yazar(self):
        return self._yazar

    @yazar.setter
    def yazar(self, value):
        if value.strip():
            self._yazar = value

    @property
    def sayfa_sayisi(self):
        return self._sayfa_sayisi

    @sayfa_sayisi.setter
    def sayfa_sayisi(self, value):
        self._sayfa_sayisi = value

    def __str__(self):
        return f"[Kitap] No: {self.kayitNo} | Başlık: {self.baslik} | Yazar: {self.yazar} | Sayfa: {self.sayfa_sayisi}"


class Dergi(Kaynak):
    """Kaynak sınıfından türetilen Dergi sınıfı."""
    def __init__(self, baslik, kayitNo, yayin_donemi, sayi_no):
        super().__init__(baslik, kayitNo)
        self._yayin_donemi = yayin_donemi
        self._sayi_no = sayi_no

    @property
    def yayin_donemi(self):
        return self._yayin_donemi

    @yayin_donemi.setter
    def yayin_donemi(self, value):
        if value.strip():
            self._yayin_donemi = value

    @property
    def sayi_no(self):
        return self._sayi_no

    @sayi_no.setter
    def sayi_no(self, value):
        self._sayi_no = value

    def __str__(self):
        return f"[Dergi] No: {self.kayitNo} | Başlık: {self.baslik} | Dönem: {self.yayin_donemi} | Sayı: {self.sayi_no}"



#  CRUD


class Islem(ABC):
    """CRUD işlemlerini zorunlu kılan soyut ana sınıf."""
    @abstractmethod
    def ekle(self):
        pass

    @abstractmethod
    def sil(self):
        pass

    @abstractmethod
    def guncelle(self):
        pass

    @abstractmethod
    def listele(self):
        pass


class KitapIslem(Islem):
    """Kitapların CRUD operasyonlarını ve listesini yöneten sınıf."""
    def __init__(self):
        self.kitaplar = []

    def toplam_kitap_sayisi(self):
        return len(self.kitaplar)

    def ekle(self):
        print("\n---** Kitap Ekleme Paneli **---")
        kayitNo = input("Kitap Kayıt Numarası: ").strip()

        # BONUS: Benzersiz kayıt numarası kontrolü
        for k in self.kitaplar:
            if k.kayitNo == kayitNo:
                print(" HATA: Bu kayıt numarasına sahip bir kitap zaten var!")
                return

        baslik = input("Kitap Adı: ").strip()
        yazar = input("Yazar Adı: ").strip()
        sayfa_sayisi = input("Sayfa Sayısı: ").strip()

        if not (kayitNo and baslik and yazar and sayfa_sayisi):
            print(" HATA: Tüm alanları doldurmak zorundasınız.")
            return

        yeni_kitap = Kitap(baslik, kayitNo, yazar, sayfa_sayisi)
        self.kitaplar.append(yeni_kitap)
        print(" Kitap başarıyla sisteme eklendi.")
        print(f" Toplam Kitap Sayısı: {self.toplam_kitap_sayisi()}")

    def sil(self):
        print("\n---** Kitap Silme Paneli **---")
        kayitNo = input("Silinecek Kitabın Kayıt Numarası: ").strip()
        for k in self.kitaplar:
            if k.kayitNo == kayitNo:
                self.kitaplar.remove(k)
                print(f" {kayitNo} numaralı kitap başarıyla silindi.")
                return
        print(" Kayıt bulunamadı.")

    def guncelle(self):
        print("\n---** Kitap Güncelleme Paneli **---")
        kayitNo = input("Güncellenecek Kitabın Kayıt Numarası: ").strip()
        for k in self.kitaplar:
            if k.kayitNo == kayitNo:
                print(f"Mevcut Veri -> Başlık: {k.baslik} | Yazar: {k.yazar} | Sayfa: {k.sayfa_sayisi}")

                yeni_baslik = input("Yeni Başlık (Pas geçmek için Enter): ").strip()
                yeni_yazar = input("Yeni Yazar (Pas geçmek için Enter): ").strip()
                yeni_sayfa = input("Yeni Sayfa Sayısı (Pas geçmek için Enter): ").strip()

                if yeni_baslik: k.baslik = yeni_baslik
                if yeni_yazar: k.yazar = yeni_yazar
                if yeni_sayfa: k.sayfa_sayisi = yeni_sayfa

                print(" Kitap bilgileri başarıyla güncellendi.")
                return
        print(" Kayıt bulunamadı.")

    def listele(self):
        print("\n---** Kayıtlı Kitaplar **---")
        # BONUS: Boş liste kontrolü
        if not self.kitaplar:
            print("Kayıt bulunamadı.")
            return
        for k in self.kitaplar:
            print(k)


class DergiIslem(Islem):
    """Dergilerin CRUD operasyonlarını ve listesini yöneten sınıf."""
    def __init__(self):
        self.dergiler = []

    def toplam_dergi_sayisi(self):
        return len(self.dergiler)

    def ekle(self):
        print("\n---** Dergi Ekleme Paneli **---")
        kayitNo = input("Dergi Kayıt Numarası: ").strip()

        # BONUS: Benzersiz kayıt numarası kontrolü
        for d in self.dergiler:
            if d.kayitNo == kayitNo:
                print(" HATA: Bu kayıt numarasına sahip bir dergi zaten var!")
                return

        baslik = input("Dergi Adı: ").strip()
        yayin_donemi = input("Yayın Dönemi (Aylık/Haftalık): ").strip()
        sayi_no = input("Sayı Numarası: ").strip()

        if not (kayitNo and baslik and yayin_donemi and sayi_no):
            print(" HATA: Tüm alanları doldurmak zorundasınız.")
            return

        yeni_dergi = Dergi(baslik, kayitNo, yayin_donemi, sayi_no)
        self.dergiler.append(yeni_dergi)
        print(" Dergi başarıyla sisteme eklendi.")
        print(f" Toplam Dergi Sayısı: {self.toplam_dergi_sayisi()}")

    def sil(self):
        print("\n---** Dergi Silme Paneli **---")
        kayitNo = input("Silinecek Derginin Kayıt Numarası: ").strip()
        for d in self.dergiler:
            if d.kayitNo == kayitNo:
                self.dergiler.remove(d)
                print(f" {kayitNo} numaralı dergi başarıyla silindi.")
                return
        print(" Kayıt bulunamadı.")

    def guncelle(self):
        print("\n---** Dergi Güncelleme Paneli **---")
        kayitNo = input("Güncellenecek Derginin Kayıt Numarası: ").strip()
        for d in self.dergiler:
            if d.kayitNo == kayitNo:
                print(f"Mevcut Veri -> Başlık: {d.baslik} | Dönem: {d.yayin_donemi} | Sayı: {d.sayi_no}")

                yeni_baslik = input("Yeni Başlık (Pas geçmek için Enter): ").strip()
                yeni_donem = input("Yeni Yayın Dönemi (Pas geçmek için Enter): ").strip()
                yeni_sayi = input("Yeni Sayı No (Pas geçmek için Enter): ").strip()

                if yeni_baslik: d.baslik = yeni_baslik
                if yeni_donem: d.yayin_donemi = yeni_donem
                if yeni_sayi: d.sayi_no = yeni_sayi

                print(" Dergi bilgileri başarıyla güncellendi.")
                return
        print(" Kayıt bulunamadı.")

    def listele(self):
        print("\n---** Kayıtlı Dergiler **---")
        # BONUS: Boş liste kontrolü
        if not self.dergiler:
            print("Kayıt bulunamadı.")
            return
        for d in self.dergiler:
            print(d)



# Menü


class Menu:
    """Kullanıcı arayüzünü ve ana döngüyü yöneten ana sınıf."""
    def __init__(self):
        self.kitap_sistemi = KitapIslem()
        self.dergi_sistemi = DergiIslem()

    def calistir(self):
        while True:
            print("\n" + "="*35)
            print(" KÜTÜPHANE YÖNETİM SİSTEMİ")
            print("="*35)
            print("1. Kitap Ekle")
            print("2. Kitap Sil")
            print("3. Kitap Güncelle")
            print("4. Kitapları Listele")
            print("5. Dergi Ekle")
            print("6. Dergi Sil")
            print("7. Dergi Güncelle")
            print("8. Dergileri Listele")
            print("9. Çıkış")
            print("="*35)

            secim = input("Yapmak istediğiniz işlemi seçiniz (1-9): ").strip()

            if secim == "1":
                self.kitap_sistemi.ekle()
            elif secim == "2":
                self.kitap_sistemi.sil()
            elif secim == "3":
                self.kitap_sistemi.guncelle()
            elif secim == "4":
                self.kitap_sistemi.listele()
            elif secim == "5":
                self.dergi_sistemi.ekle()
            elif secim == "6":
                self.dergi_sistemi.sil()
            elif secim == "7":
                self.dergi_sistemi.guncelle()
            elif secim == "8":
                self.dergi_sistemi.listele()
            elif secim == "9":
                print("\nSistemden güvenli bir şekilde çıkış yapıldı. İyi çalışmalar dilerim!")
                break
            else:
                print(" GEÇERSİZ SEÇİM! Lütfen 1 ile 9 arasında bir rakam girin.")


# Programın başlatılması
if __name__ == "__main__":
    otomasyon = Menu()
    otomasyon.calistir()
```
